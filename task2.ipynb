## Функция отображения данных для выборки

```python
import matplotlib.pyplot as plt
import numpy as np

def represent_sample(frame):
    bmi = frame["bmi"]

    mean = bmi.mean()
    print("Выборочное среднее: ", mean)

    var = bmi.var(ddof=0)
    print("Выборочная дисперсия: ", var)

    median = bmi.median()
    print("Выборочная медиана: ", median)

    bmi.hist(bins=15)
    plt.title("Гистограмма BMI")
    plt.xlabel("BMI")
    plt.ylabel("Число людей")
    plt.show()

    bmi.plot.box()
    plt.title("Boxplot BMI")
    plt.xticks([])
    plt.ylabel("BMI")

    plt.show()

    sorted_bmi = sorted(bmi)
    pd.DataFrame({0: sorted_bmi, 1:np.arange(len(sorted_bmi)) / len(sorted_bmi)}).plot(0, 1, legend=None)
    plt.title("Эмпирическая функция распределения")
    plt.xlabel("BMI")
    plt.ylabel("Вероятность")
    plt.show()
```

## Выборка со всеми людьми

```python
import pandas as pd

people = pd.read_csv("sex_bmi_smokers.csv")
represent_sample(people)
```

## Выборка курильщика

```python
smoker = people[people["smoker"] == "yes"]
represent_sample(smoker)
```

## Выборка здорового человека

```python
non_smoker = people[people["smoker"] == "no"]
represent_sample(non_smoker)
```
